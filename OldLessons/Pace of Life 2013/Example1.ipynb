### Time to analyze

I would like to analyze the relationships between the DASS total and whether students eat dinner in the DH.

```r
ggplot(pol, aes(x = DHDinner, y = DassTotal)) +
  geom_boxplot()+
    geom_jitter(width = 0.25)+
    xlab("Responses to Dining Hall Dinner")+
    ylab("DASS Stress level")+
    ggtitle("Students that miss dinner have more stress.")
```

## Analysis

These box plots show that the median amount of stress of student that misses dinner is higher than the third quartile otherwise. This suggests that having dinner in the dining hall with friends is an important experience to bring students back to equillibrium.
